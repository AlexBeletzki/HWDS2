```python
t = np.linspace(0, 10, num = 51)
t
```

```python
f = np.cos(t)
f
```

```python
plt.plot(t, f, color = 'Green')
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.axis = ([0.5, 9.5, -2.5, 2.5])

plt.show()
```

```python
x = np.linspace(-3, 3, num = 51)
x
```

```python
y1 = x ** 2
y1
```

```python
y2 = 2 * x + 0.5
y2
```

```python
y3 = -3 * x - 1.5
y3
```

```python
y4 = np.sin(x)
y4
```

```python
fig = plt.figure()
fig, ax = plt.subplots(nrows=2, ncols=2)

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace = 0.3)

ax1, ax2, ax3, ax4 = ax.flatten()


ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
ax4.plot(x, y4)

ax1.set(title = 'График 1')
ax2.set(title = 'График 2')
ax3.set(title = 'График 3')
ax4.set(title = 'График 4')

ax1.set_xlim([-5, 5])
```
